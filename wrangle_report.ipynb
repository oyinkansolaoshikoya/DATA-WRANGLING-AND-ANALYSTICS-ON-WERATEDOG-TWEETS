## Reporting: wragle_report


The wrangling process began by gathering data from three different data sources which were “twitter-archive-enhanced.csv “ ,  “image_predictions.tvc”, and “tweet_json.txt” files. The dataset focused on tweets’ related to dogs’ rating, favorite tweets, re-tweet, dogs' stage and prediction which had following wrangling steps carried out sequentially 

#### 1.Gathering: This process involved gathering data from the different sources in which some columns were extracted from the “tweet_json.txt” for easy data analytics. 


 
#### 2.Assessing: The step involved investigating the quality issue and tidiness of the dataset through the use of visual and programmatic method such as using the following command lines df.head() and df.describe() respectively. The following quality issues were detected:




-The "tweet_id" column on "twitter-archive-enhanced" table should be an object string not an integer.

-The "Timestamp" column on "twitter-archive-enhanced" table is a string instead of a date type.

-Some values in "rating_numerator" and "rating_denominator" on "twitter-archive-enhanced" table are outliers because values are out of range.

-There are dogs with wrong name on the "name" column on the "twitter-archive-enhanced" table which should be replace with "None".

-There are re-tweets mixed up with original tweets based on "RT @" in the value of the "text" column.

-The columns labelled "in_reply_to_status_id","in_reply_to_user_id", "retweeted_status_id", "retweeted_status_user_id", and "retweeted_status_timestamp" on "twitter-archive-enhanced" have a lot of missing values. As a result, it will not be relevant for analytics.

-The "tweet_id" on "image_predictation" table should be an object string not an integer.

-There are duplicate urls on the "jpg_url" column on the "image_prediction" table for the same image.

-The"tweet_id" on "tweet-json.txt"" table should be an object string not an integer.

-The column labelled 'created_at' on "tweet-json.txt" table is an object instead of a date type.


##### While the tidiness issue were:

-The columns labelled "name", "doggo","floofer","pupper", and "puppo" on "twitter-archive-enhanced" table needs to be normalised to one column because it has similar attributes.

-The columns labelled "p2", "p2_conf",	"p2_dog", "p3", "p3_conf",	"p3_dog" on "image_predictation" table will be dropped to only use some columns for prediction.

-The columns names will be renamed from "p1", "p1_conf", and "p1_dog" to "prediction", "confidence_level", and "validation" for appropriate communication.






#### 3. Cleaning: After detecting the quality and tidiness issue above, data was cleaned by defining the issues, writing codes to remove or fixed the different issues and testing to confirm that the issues do no longer exist on the various sources which was eventually merged and stored as "twitter_archive_master.csv"
